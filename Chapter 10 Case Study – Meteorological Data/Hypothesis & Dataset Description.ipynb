```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('ancona_weather.csv')
c = pd.read_csv('cremona_weather.csv')

plt.plot(a['DATE'], a['TAVG'], label='Ancona', color='blue')
plt.plot(c['DATE'], c['TAVG'], label='Cremona', color='red')

plt.title('Avg Temp: Coastal vs Inland')
plt.xlabel('Date')
plt.ylabel('TAVG (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
