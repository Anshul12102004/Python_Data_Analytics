```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Two Arrows.png")
if img is None:
    print("Image not found.")
else:
 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    canny = cv2.Canny(gray, 100, 200)
    
    # Calculate combined Sobel (without storing intermediates)
    sx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    sy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    combined = cv2.addWeighted(
        cv2.normalize(sx, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U),
        0.5,
        cv2.normalize(sy, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U),
        0.5, 0
    )
    plt.figure(figsize=(10, 8))
    
    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original')
    plt.axis('off')
    
    plt.subplot(2, 2, 2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')
    
    plt.subplot(2, 2, 3)
    plt.imshow(canny, cmap='gray')
    plt.title('Canny')
    plt.axis('off')
    
    plt.subplot(2, 2, 4)
    plt.imshow(combined, cmap='gray')
    plt.title('Sobel Combined')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```
