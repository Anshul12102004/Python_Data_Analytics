```python
from IPython.display import display, HTML

# D3.js grouped bar chart code
d3_code = """
<div id="chart"></div>
<script src="https://d3js.org/d3.v7.min.js"></script>
<script>
// Wait for D3 to load and DOM to be ready
setTimeout(function() {
  const data = [
    {day: "Mon", Ancona: 9, Cremona: 5},
    {day: "Tue", Ancona: 10, Cremona: 4}
  ];

  // Create a unique ID to avoid conflicts with multiple charts
  const chartId = "chart-" + Math.random().toString(36).substr(2, 9);
  d3.select("#chart").attr("id", chartId);

  const svg = d3.select("#" + chartId)
    .append("svg")
    .attr("width", 300)
    .attr("height", 150);

  const x0 = d3.scaleBand()
    .domain(data.map(d => d.day))
    .range([0, 280])
    .padding(0.2);

  const x1 = d3.scaleBand()
    .domain(["Ancona", "Cremona"])
    .range([0, x0.bandwidth()])
    .padding(0.05);

  const y = d3.scaleLinear()
    .domain([0, 12])
    .range([120, 0]);

  svg.append("g")
    .attr("transform", "translate(30,20)")
    .selectAll("g")
    .data(data)
    .enter().append("g")
      .attr("transform", d => `translate(${x0(d.day)},0)`)
      .selectAll("rect")
      .data(d => ["Ancona", "Cremona"].map(key => ({key: key, value: d[key]})))
      .enter().append("rect")
        .attr("x", d => x1(d.key))
        .attr("y", d => y(d.value))
        .attr("width", x1.bandwidth())
        .attr("height", d => 120 - y(d.value))
        .attr("fill", d => d.key === "Ancona" ? "steelblue" : "tomato");
}, 500); // Delay to ensure D3 is loaded
</script>
"""

# Display the HTML with D3 code
display(HTML(d3_code))
```
